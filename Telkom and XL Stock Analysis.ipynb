# Daily Percentage Change

Here we want to know how much return did we get if we sell our stock next day. Is it positive or negative. The larger return we got daily it means that the stock is volatile. This analysis can determine wether a stock stable or not. It is important for investing strategy

rt = pt/p(t-1) - 1

```python
telkom_data['Return'] = (telkom_data['Close']/telkom_data['Close'].shift(1))-1
```

```python
telkom_data
```

```python
xl_data['Return'] = (xl_data['Close']/xl_data['Close'].shift(1))-1
```

```python
xl_data
```

```python
telkom_data['Return'].hist(bins=50, figsize=(15,7))
```

## XL Return Daily

```python
xl_data['Return'].hist(bins=50, figsize=(15,7))
```

The fatter our histogram, the more volatile. The width of those bell shape related to standar deviation of the distribution. The higher it is, the more variance. It related to standar deviation. The higgher standar deviation, the higher variance, the more volatile the stock

# Cummulative Return

Cumulative return compute return relative to the day investment has been made. It will give investor insight on a long-term investment. If I invested $1 the first day in the company, how much would it return today. If cumulative return above 1, we're making provit

it = (1+rt)i(t-1) 

```python
telkom_data['Cumulative Return']=(1+telkom_data['Return']).cumprod()
xl_data['Cumulative Return']=(1+xl_data['Return']).cumprod()
```

```python
telkom_data['Cumulative Return'].plot(label='Telkom',figsize=(15,7))
xl_data['Cumulative Return'].plot(label='XL')
plt.legend()
plt.title('Cumulative Return Vs Time')
plt.show()
```
